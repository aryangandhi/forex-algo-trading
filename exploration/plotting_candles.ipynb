```python
import pandas as pd
import plotly.graph_objects as go
import datetime as dt
```

```python
pair = 'EUR_USD'
granularity = "H4"
df = pd.read_pickle(f'../data/{pair}_{granularity}.pkl')
```

```python
df["sTime"] = [dt.datetime.strftime(x,"s%Y-%m-%d %H:%M") for x in df.time]
```

```python
df.tail()
```

```python
df_plot = df.iloc[:100]
```

```python
fig = go.Figure()
fig.add_trace(go.Candlestick(
    x = df_plot.sTime,
    open = df_plot.mid_o,
    high = df_plot.mid_h,
    low = df_plot.mid_l,
    close = df_plot.mid_c,
    line=dict(width=1), 
    opacity=1,
    
))

fig.update_yaxes(
gridcolor='#1f292f'

)

fig.update_xaxes(
    gridcolor='#1f292f',
    nticks=10
)

fig.update_layout(
    width=1000,
    height=400,
    margin=dict(l=10,r=10,b=10,t=10),
    paper_bgcolor='#2c303c',
    plot_bgcolor='#2c303c',
    font=dict(size=8, color='#e1e1e1')
)

fig.show()
```
